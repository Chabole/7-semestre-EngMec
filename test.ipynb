```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Importando e tratando
df_2 =pd.read_excel('PROJECAO_RODANTES_2.xlsx')

df_2 = df_2.drop(columns=['Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'FROTA', 'LE'])
df_2['DATA'] = pd.to_datetime(df_2['DATA'])
df_2.set_index('DATA', drop=True, inplace=True)

#Visualizando os dados
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(df_2.index, df_2.LD)
```

```python
#Separando
x_test, x_train, y_test, y_train = train_test_split(df_2.index, df_2.LD)

#Visualizando a separação dos dados
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(x_test, y_test, color='r', label='Test data')
ax.scatter(x_train, y_train, color='b', label='Train data')
ax.grid(ls='dotted')
ax.legend()
```

```python
#Criando modelo
modelo = LinearRegression()
modelo.fit(x_train.values.reshape(-1,1), y_train.values.reshape(-1,1))
```
