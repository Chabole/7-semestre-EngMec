```python
import pandas as pd
import matplotlib.pyplot as plt
import ZebraLib as zb

df = pd.read_excel('D:/UNESP/7 semestre - Eng/Lab. TCM I/Dados_exp7_copo.xlsx', sheet_name='Planilha2')

t = df['tempo_s'] 
Ts = df['Ts'] 
Td = df['Td'] 
Tt = df['Tt'] 

ps = zb.fit(t, Ts, 6)
pd = zb.fit(t, Td, 6)
pt = zb.fit(t, Tt, 6)

fig, ax = plt.subplots()
ax.set(xlabel='Tempo[segundos]', ylabel='Temperatura[ºC]')

ax.scatter(t, Ts, color='red', label='Copo com parede simples')
ax.plot(t, ps(t), linestyle='dotted',color='red')

ax.scatter(t, Td, color='blue', label='Copo com parede dupla')
ax.plot(t, pd(t), linestyle='dotted',color='blue')

ax.scatter(t, Tt, color='orange', label='Copo com parede tripla')
ax.plot(t, pt(t), linestyle='dotted',color='orange')

ax.grid(linestyle='dotted')
ax.legend()

fig.savefig('D:/trab.pdf', bbox_inches='tight')
```
