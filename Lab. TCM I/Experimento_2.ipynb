```python
E, I, qx, k32, k42 = [], [], [], [],[]
for i in range(len(En_value)):
    E.append(ufloat(En_value[i], 0.1/np.sqrt(3)))
    I.append(ufloat(In_value[i], 0.1/(np.sqrt(3))))
    qx.append((E[i]*I[i])/2)

    k32.append((qx[i]*L)/(A*(T3[i] - T1[i])))
    k42.append((qx[i]*L)/(A*(T4[i] - T2[i])))
```
