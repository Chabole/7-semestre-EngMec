```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import ZebraLib as zb
from scipy.integrate import quad, solve_ivp
import scipy.io

#Dado
df_dados = scipy.io.loadmat('D:/UNESP/EngMec-UNESP/Controle Sistemas/Laboratórios/Lab 4 - Malhas/curves_again.mat')
df_dados['curves']
```

```python
#colum é é aposição para qual variavel "pegar"
def extrair_data(colum):
    data_0, data_1 = [], []
    for i in range(df_dados['curves'].shape[0]):
        data_0.append(df_dados['curves'][i][colum][0]) #medida1
        data_1.append(df_dados['curves'][i][colum][1]) #medida2
    return data_0, data_1

#pegando todos os dados
e1, e2 = extrair_data(0)
sup1, sup2 = extrair_data(1)
inf1, inf2 = extrair_data(2)
T1, T2 = extrair_data(3)
t1, t2 = extrair_data(4)
```

```python
#%matplotlib
fig, ax = plt.subplots()
ax.set(xlabel='Tempo(s)', ylabel='Temperatura(°C)', ylim=(28,34))

ax.plot(t1, e1, label='Referência')
ax.plot(t1, sup1, label='Limite superior')
ax.plot(t1, inf1, label='Limite inferior')
ax.plot(t1, T1, label='Temperatura')

ax.legend()
ax.grid(ls='dotted')

fig.savefig('res1.pdf', bbox_inches='tight', dpi=600)
```

```python
fig, ax = plt.subplots()
ax.set(xlabel='Tempo(s)', ylabel='Temperatura(°C)', ylim=(29,37))

ax.plot(t2, e2, label='Referência')
ax.plot(t2, sup2, label='Limite superior')
ax.plot(t2, inf2, label='Limite inferior')
ax.plot(t2, T2, label='Temperatura')
ax.legend()
ax.grid(ls='dotted')

fig.savefig('res2.pdf', bbox_inches='tight', dpi=600)
```
