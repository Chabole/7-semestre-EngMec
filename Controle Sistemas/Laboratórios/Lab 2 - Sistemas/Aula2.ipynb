```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import ZebraLib as zb
from scipy.integrate import quad, solve_ivp
import scipy.io

cp = 4200
C =200
ta = 25
Tent = 30
qent = 5000

qsai = lambda T, R: (ta - T) / R

m = 0.05
R = [np.inf, 0.05]

t = np.arange(0, 10, .10)
T0= Tent

Ts = np.zeros([len(R), len(t)])
```

```python
#%matplotlib
fig, ax = plt.subplots()
ax.set(xlabel='Tempo(s)', ylabel='Temperatura')
for Ri in R:
    dTdt = lambda t, T: (1/C)*(m*cp*(Tent - T) + qent + qsai(T, Ri))
    
    sol = solve_ivp(dTdt, t_span=(0, 10), y0=[T0], t_eval=t)
    ax.plot(sol.t, sol.y[0], label=f'R={Ri}')
    ax.legend()
ax.grid(ls='dotted')
```
