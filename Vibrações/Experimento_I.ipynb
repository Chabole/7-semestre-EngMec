```python
Y1_viga = np.array([29.449, 15.614, 290.897, 36.299, 69.251, 118.733, 49.535, 47.285, 27.269, 135.363])
Y2_viga = np.array([26.480, 18.875, 261.772, 30.786, 69.389, 93.930, 49.451, 34.627, 29.232, 99.440])

t1 = np.array([0.445, 0.469, 0.488, 0.415, 0.412, 0.416, 0.377, 0.361, 0.376, 0.369])
t2 = np.array([0.511, 0.531, 0.553, 0.484, 0.475, 0.483, 0.440, 0.428, 0.436, 0.436])
ra = np.array([0.066, 0.062, 0.065, 0.069, 0.063, 0.067, 0.063, 0.067, 0.060, 0.067])
```

```python
T = t2 - t1
desvio_1 = np.log(Y1_viga/Y2_viga)
amortec = desvio_1/(desvio_1**2 + 4*np.pi**2)
Wn = 2*np.pi/(T*np.sqrt(1-amortec**2))
Wa = 2*np.pi/T

pd.DataFrame({'X1':Y1_viga, 'X2':Y2_viga, 'desvio_1':desvio_1, 'amortec':amortec, 'Wn':Wn, 'Wa':Wa})
```

## Teórico - Comparação

```python
#Dados
L= 20.1/100
h_values= np.array([31.5, 31.7, 31.4, 31.4, 31.35])
e_values=  np.array([2, 2, 1.95, 1.9, 2.00])

#Incertezas do tipo A - h e espessura
h = ufloat(h_values.mean(), h_values.std(ddof=1)/np.sqrt(len(h_values)))
e = ufloat(e_values.mean(), e_values.std(ddof=1)/np.sqrt(len(e_values))) 

I = (h*e**3)/12
I = I/1000**4

k = (3*(200e9)*I)/L**3
```

```python
#considerando a massa do acelerometro
#teORICO
wat = (k/(0.11+0.02))**0.5
wat = wat/(2*3.1415)

#eXPERIMENTal
W = Wn/(2*np.pi)
Wax= ufloat(W.mean(), W.std(ddof=1)/np.sqrt(len(W)))

(Wax - wat)/wat*100
```

```python
#NÃO considerando a massa do acelerometro
wat = (k/(0.11))**0.5
wat = wat/(2*3.1415)

W = Wn/(2*np.pi)
Wax= ufloat(W.mean(), W.std(ddof=1)/np.sqrt(len(W)))

(Wax - wat)/wat*100
```
