```python
setor='Intermediários Financeiros'
group = df.groupby(by='Subsetor de Atuação')
lista_sort = group.get_group(setor).sort_values(by=['P/L'], ascending=False)
lista_sort
```

```python
df_sorted = pd.DataFrame()
for setor in df['Subsetor de Atuação'].unique():
    group = df.groupby(by='Subsetor de Atuação')

    lista_sort = group.get_group(setor).sort_values(by=['P/VP'], ascending=False)[:5]
    df_sorted = df_sorted.append(lista_sort)

df_sorted = df_sorted.reset_index(drop=True)
#df_sorted[df_sorted['Subsetor de Atuação']=='Máquinas e Equipamentos']
df_sorted
```
