```python
import plotly.io as pio
#pio.renderers.default = "plotly_mimetype"
#pio.renderers.default = "kaggle"
#pio.renderers.default = "browser"
import pandas as pd
import time 
import json
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
from helps import *

stocks = pd.read_excel('D:/UNESP/Finance/outputs.xlsx', index_col=0)
stocks = stocks[(stocks['VOLUME (dia)'] >=1e6) & (stocks['P/L']>0) & (stocks['P/L']<100)]

stocks = stocks.drop(columns=['Valor atual', 'Min. 52 semanas',
       'Máx. 52 semanas', 'Valorização (12m)', 'TOMADOR (média)',
       'DOADOR (média)', 'Nº DE AÇÕES ALUGADAS (dia)',
       'Nº DE CONTRATOS', 'Unnamed: 14', 'Ano passado', 'Ano atual',
       'Provisionado', 'Ativos', 'Ativo circulante', 'Dívida bruta',
       'Disponibilidade', 'Dívida líquida', 'Valor de mercado',
       'Valor de firma', 'Segmento de listagem', 'Free Float', 'M. Bruta',
       'M. EBITDA', 'M. EBIT', 'Giro ativos', 'UNIT - Composição', 'EV/EBITDA',
       'EV/EBIT', 'P/EBITDA', 'P/EBIT', 'VPA', 'P/Ativo', 'P/SR',
       'P/Cap. Giro', 'P/Ativo Circ. Liq.', 'Dív. líquida/PL',
       'Dív. líquida/EBITDA', 'VOLUME (dia)', 'PL/Ativos',	'Dív. líquida/EBIT', 'Passivos/Ativos', 'Segmento de Atuação', 'Setor de Atuação'])

#df['Subsetor de Atuação'].unique()
#df = compare_df(stocks)
stocks['Pontos'] = (stocks['ROE']*stocks['LPA']*stocks['D.Y'])/(stocks['P/L']*stocks['P/VP'])

df_best_points = stocks.sort_values(by='Pontos', ascending=False).head(50)
df_best_points
```

```python
df_best_points.sort_values(by=['Subsetor de Atuação', 'Pontos'], ascending=False)
```

```python
fig = px.scatter(df_best_points, x='Pontos', y='Subsetor de Atuação', color="P/VP", height=600,
                 size="D.Y", hover_data=['Ativo', 'LPA', 'ROE', 'P/L', 'ROIC', 'P/VP'], 
                color_continuous_scale= px.colors.sequential.Jet_r)
fig.show()   
```
