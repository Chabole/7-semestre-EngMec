```python
lista_valores = []
#ativos = ['MGLU3', 'ABEV3', 'AALR3']
ativos = stocks['Ativo'].values

for ativo in ativos:

    seg = stocks[stocks['Ativo']==ativo]['Subsetor de Atuação'].values[0]
    df = stocks.groupby(by='Subsetor de Atuação')
    res = df[['P/VP', 'P/L', 'D.Y', 'LPA', 'ROE']].mean()
    categ = res.loc[seg]
    
    at = stocks[stocks['Ativo'] == ativo][['Ativo', 'P/VP', 'P/L', 'D.Y', 'LPA', 'ROE']]

    try:
        valor = (float(categ['P/VP']) - float(at['P/VP']))/float(categ['P/VP'])
    except:
        valor = 69.69
    lista_valores.append(valor)

calc = pd.DataFrame({'Sub':stocks['Subsetor de Atuação'],'Ativos':stocks['Ativo'].values, 'P/VP':lista_valores})
calc
```

```python
calc = calc[(calc['P/VP'] > 0) & (calc['P/VP'] != 69.69) ]
df_sort = calc.sort_values(by=['P/VP']) #agrupar por subsetor
df_sort
```

```python
fig = px.scatter(df_sort, x="P/VP", y="Sub", color="Sub", hover_data=['Ativos'])
fig.show()   
```
